```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
proc_cnt = np.array([i for i in range(1, 5)])
time_thousand = np.array([0.0554819, 0.0309901, 0.0214431, 0.0167878])
acceleration_thousand = time_thousand[0] / time_thousand

plt.xlabel('Number of processes')
plt.ylabel('Acceleration')
plt.title('2000 points')
plt.plot(proc_cnt, acceleration_thousand)
plt.scatter(proc_cnt, acceleration_thousand)
plt.plot(proc_cnt, [i for i in range(1, 5)])
plt.show()
```

```python
plt.clf()
time_million = np.array([6.74944, 3.75726, 2.50852, 1.82427])
acceleration_million = time_million[0] / time_million

plt.xlabel('Number of processes')
plt.ylabel('Acceleration')
plt.title('10000 points')
plt.plot(proc_cnt, acceleration_million)
plt.scatter(proc_cnt, acceleration_million)
plt.plot(proc_cnt, [i for i in range(1, 5)])
plt.show()
```
