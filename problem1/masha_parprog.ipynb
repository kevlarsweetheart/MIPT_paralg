```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
proc_cnt = np.array([i for i in range(1, 13)])
time_thousand = np.array([7.39098e-05, 0.000123978, 0.000154972, 0.000584126, 0.00077486, 0.00122499, 0.00113201, 0.00166607, 0.00171685, 0.00200987, 0.00227499, 0.00279212])
acceleration_thousand = time_thousand[0] / time_thousand

plt.xlabel('Number of processes')
plt.ylabel('Acceleration')
plt.plot(proc_cnt, acceleration_thousand)
plt.scatter(proc_cnt, acceleration_thousand)
plt.plot(proc_cnt, [i for i in range(1, 13)])
plt.show()
```
